## Задание 1

```python
# https://www.matburo.ru/tvart_sub.php?p=art_moneta
from scipy.special import factorial
import numpy as np

prime_numbers = [2, 3, 5, 7, 11, 13, ]
n=15
```

### Способ №1

```python
sum(factorial(n)/
    (factorial(np.array([n]*len(prime_numbers))-np.array(prime_numbers))
                *factorial(prime_numbers)
                *2**n
                ))
```

### Способ №2

```python
def formula_bernulli(n, k, p=0.5):
    C_k_n = factorial(n)/(factorial(n-k)
                        *factorial(k)
                        )
    return C_k_n*p**k*(1-p)**(n-k)
```

```python
sum([formula_bernulli(n,_) for _ in prime_numbers])
```

## Задание 2

```python
import random
from matplotlib import pyplot as plt
import seaborn as sns

cm = ['r', 'g', 'b', 'aqua', 'yellow']
title = ['Нормальное', 'Экспоненциальное', 'Стьюдента']

norm =[np.random.normal(0, 0.1, 1000),
       np.random.normal(10, 5, 900),
       np.random.normal(50, 10, 700),
       np.random.normal(100, 15, 500),
       np.random.normal(150, 20, 300),
]

exp =[np.random.exponential(scale=0.2, size=1500),
       np.random.exponential(scale=0.7, size=1200),
       np.random.exponential(scale=1.0, size=900),
       np.random.exponential(scale=1.2, size=600),
       np.random.exponential(scale=2, size=300),
]

student =[np.random.standard_t(10, 1500),
       np.random.standard_t(20, 1200),
       np.random.standard_t(30, 900),
       np.random.standard_t(100, 600),
       np.random.standard_t(150, 300),
]
```

```python
def drow_graph(*distribution, title, cm=cm):  
    plt.figure(figsize=(15,7))
    for i, _ in enumerate(distribution):
        sns.histplot(x=_, bins=25, color=cm[i], label=title+"_"+str(i+1), kde=True)
    plt.title(title)
    plt.legend() 
    plt.show()
```

```python
drow_graph(*norm, title=title[0])
```

```python
drow_graph(*exp, title=title[1])
```

```python
drow_graph(*student, title=title[2])
```
