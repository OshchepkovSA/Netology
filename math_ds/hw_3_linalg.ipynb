Дан объект в 2d пространстве:



```python
import numpy as np
import matplotlib.pyplot as plt
A = [[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]]
A = np.array(A)
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

При помощи линейных отображений:

● Уменьшить объект в два раза

● Повернуть на 130 градусов

● Отразить объект относительно прямой y=x

● Вычислить определитель композиции преобразований

(итоговой матрицы реализующей все три отображения

 Уменьшить объект в два раза

```python
#╰( ͡° ͜ʖ ͡° )つ──☆*:・ﾟ
# Если имелось ввиду уменьшить по оси х и у в 2 раза то решение ниже.
I = np.eye(3)
T1 = np.array([0.5, 0.5, 1])
T1 = I*T1
A1 = T1@A.T
A1 = A1.T

x = A1[:,0]
y = A1[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Повернуть на 130 градусов

```python
#╰( ͡° ͜ʖ ͡° )つ──☆*:・ﾟ
# Исхожу из того, что поворот требуется по часовой стрелке
# https://coderoad.ru/57648391/%D0%9A%D0%B0%D0%BA-%D0%BF%D0%BE%D0%B2%D0%B5%D1%80%D0%BD%D1%83%D1%82%D1%8C-%D0%B8%D0%B7%D0%BE%D0%B1%D1%80%D0%B0%D0%B6%D0%B5%D0%BD%D0%B8%D0%B5-%D0%B2%D1%80%D1%83%D1%87%D0%BD%D1%83%D1%8E-%D0%B1%D0%B5%D0%B7-%D0%B8%D1%81%D0%BF%D0%BE%D0%BB%D1%8C%D0%B7%D0%BE%D0%B2%D0%B0%D0%BD%D0%B8%D1%8F-cv2-getRotationMatrix2D
rotation_amount_degree = 130
rotation_amount_rad = rotation_amount_degree * np.pi / 180.0
T2 = np.array([
            [np.cos(rotation_amount_rad), np.sin(rotation_amount_rad)],
            [-np.sin(rotation_amount_rad), np.cos(rotation_amount_rad)],
            ])

A2 = T2@A1[:, 0:2].T
A2 = A2.T

x = A2[:,0]
y = A2[:,1]
plt.plot(x, y)
plt.ylim([-100, 100])
plt.xlim([-100, 100])
```

```python
#╰( ͡° ͜ʖ ͡° )つ──☆*:・ﾟ
# Исхожу из того, что поворот требуется по часовой стрелке
# https://coderoad.ru/57648391/%D0%9A%D0%B0%D0%BA-%D0%BF%D0%BE%D0%B2%D0%B5%D1%80%D0%BD%D1%83%D1%82%D1%8C-%D0%B8%D0%B7%D0%BE%D0%B1%D1%80%D0%B0%D0%B6%D0%B5%D0%BD%D0%B8%D0%B5-%D0%B2%D1%80%D1%83%D1%87%D0%BD%D1%83%D1%8E-%D0%B1%D0%B5%D0%B7-%D0%B8%D1%81%D0%BF%D0%BE%D0%BB%D1%8C%D0%B7%D0%BE%D0%B2%D0%B0%D0%BD%D0%B8%D1%8F-cv2-getRotationMatrix2D
rotation_amount_degree = 130
rotation_amount_rad = rotation_amount_degree * np.pi / 180.0
T2 = np.array([
            [np.cos(rotation_amount_rad), np.sin(rotation_amount_rad), 0],
            [-np.sin(rotation_amount_rad), np.cos(rotation_amount_rad), 0],
            [0, 0, 1]
            ])

A2 = T2@A1.T
A2 = A2.T

x = A2[:,0]
y = A2[:,1]
plt.plot(x, y)
plt.ylim([-100, 100])
plt.xlim([-100, 100])
A2
```

Отразить объект относительно прямой y=x

```python
#╰( ͡° ͜ʖ ͡° )つ──☆*:・ﾟ
# https://ru.stackoverflow.com/questions/1268854/%D0%9A%D0%B0%D0%BA-%D0%B4%D0%B8%D0%B0%D0%B3%D0%BE%D0%BD%D0%B0%D0%BB%D1%8C-%D0%B7%D0%B0%D0%BF%D0%BE%D0%BB%D0%BD%D0%B8%D1%82%D1%8C-%D0%BD%D1%83%D0%BB%D1%8F%D0%BC%D0%B8-%D1%8D%D0%BB%D0%B5%D0%BC%D0%B5%D0%BD%D1%82%D1%8B-%D0%B2%D1%8B%D1%88%D0%B5-%D0%B3%D0%BB%D0%B0%D0%B2%D0%BD%D0%BE%D0%B9-%D0%B4%D0%B8%D0%B0%D0%B3%D0%BE%D0%BD%D0%B0%D0%BB%D0%B8-%D0%B7%D0%B0%D0%BF%D0%BE%D0%BB%D0%BD%D0%B8%D1%82%D1%8C-1-%D0%BD
# T = np.ones((3, 3))
# T[np.diag_indices(T.shape[0])] = 0
T3 = np.array([[0, 1, 0],
              [1, 0, 0],
              [0, 0, 1]
              ])

A3 = T3@A2.T
A3 = A3.T

x = A3[:,0]
y = A3[:,1]
plt.plot(x, y)
plt.ylim([-100, 100])
plt.xlim([-100, 100])
T3
```

Вычислить определитель композиции преобразований

```python
#╰( ͡° ͜ʖ ͡° )つ──☆*:・ﾟ
TT = T1@T2@T3
np.linalg.det(TT)
```
