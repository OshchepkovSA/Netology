```python
import numpy as np

A1 = np.random.random([5,5])
A1
```

```python
A1.T
```

```python
np.linalg.inv(A1)
```

```python
I = np.eye(4)
A2 = np.random.random([4,4])
I*A2
```

```python
z = np.zeros([3, 3])
z
```

```python
print(np.linalg.det(A1))
print(np.linalg.matrix_rank(A1))
print(np.linalg.eig(A1))
```

```python
print(np.linalg.det(A1.T))
print(np.linalg.matrix_rank(A1.T))
print(np.linalg.eig(A1.T))
```

```python
print(np.linalg.det(np.linalg.inv(A1)))
print(np.linalg.matrix_rank(np.linalg.inv(A1)))
print(np.linalg.eig(np.linalg.inv(A1)))
```

```python
print(np.linalg.det(I))
print(np.linalg.matrix_rank(I))
print(np.linalg.eig(I))
```

```python
print(np.linalg.det(I*A2))
print(np.linalg.matrix_rank(I*A2))
print(np.linalg.eig(I*A2))
```

```python
print(np.linalg.det(z))
print(np.linalg.matrix_rank(z))
print(np.linalg.eig(z))
```
