#Постройте график функции, первой и второй производной:

$f(x)=sin^5(2x+1)$

На отрезке [-5; 5].

##Решение

$f^\prime(x)=10sin^4(2x+1)cos(2x+1)$

$f^{\prime\prime}(x)= 10sin^4(2x+1)^{\prime}cos(2x+1)+10sin^4(2x+1)cos(2x+1)^\prime=$

$=80sin^3(2x+1)cos(2x+1)cos(2x+1)-20sin^4(2x+1)sin(2x+1)=$

$=20sin^3(2x+1)(4cos^2(2x+1)-sin^2(2x+1)$

```python
def first_deriv_func_y(*x):
    """
    Первая производная
    """
    result = []
    for el in x:
        y = 10*np.sin(2*el+1)**4*np.cos(2*el+1)
        result.append(y)
    return result

def second_deriv_func_y(*x):
    """
    Вторая производная
    """
    result = []
    for el in x:
        y = 20*np.sin(2*el+1)**3*(4*np.cos(2*el+1)**2-np.sin(2*el+1)**2)
        result.append(y)
    return result
```

```python
x = np.linspace(-5, 5, 100)
y_1 = first_deriv_func_y(*x)
y_2 = second_deriv_func_y(*x)

plt.figure()
plt.rcParams['figure.figsize']=(10,5)
plt.rcParams['axes.titlesize']=(22)
plt.rcParams['axes.labelsize']=(22)


plt.plot(x, y_1, label='first derivative')
plt.plot(x, y_2, label='second derivative')
plt.title('Производная функции $f(x)=sin^5(2x+1)$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
