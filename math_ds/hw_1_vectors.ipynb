Даны векторы x = (1,1), y = (2, 0) и z = (0,2)
Необходимо в Python:

●Изобразить векторы на экране

●Изобразить точку x + y + z

●Найти угол между векторами x и y


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,1])
y = np.array([2,0])
z = np.array([0,2])

fig = plt.figure()
plt.plot(x[0], x[1], 'rx')
plt.plot(y[0], y[1], 'gx')
plt.plot(z[0], z[1], 'bx')
plt.plot(x[0]+y[0]+z[0], x[1]+y[1]+z[1], 'bo')
plt.show()

print(np.rad2deg(np.arccos((x[0]*y[0]+x[1]*y[1])/((x[0]**2+x[1]**2)**0.5*(y[0]**2+y[1]**2)**0.5))))
print(np.rad2deg(np.arccos(np.dot(x,y)/(np.linalg.norm(x)*np.linalg.norm(y)))))

```

Необходимо в Python:

f (x) = $3x(x^2+x)^3$

Посчитайте значение первой и второй производной функции

В точке x = 1

```python
import sympy as sym

x = sym.Symbol('x')
3*x*(x**2+x)**3
```

```python
# First derivative
res = sym.diff(3*x*(x**2+x)**3, x, 1)
res
```

```python
res.subs(x,1)
```

```python
# Second derivative
res = sym.diff(3*x*(x**2+x)**3, x, 2)
res
```

```python
res.subs(x,1)
```
