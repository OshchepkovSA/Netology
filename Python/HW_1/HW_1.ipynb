**Задание 1**

Даны 2 переменных, в которых хранятся строки произвольной длины: phrase_1 и phrase_2.
Напишите код, который проверяет какая из этих строк длиннее.

```python
phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
phrase_2 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)1'
if len(phrase_1) > len(phrase_2):
  print('Фраза 1 длиннее фразы 2')
elif len(phrase_1) < len(phrase_2): 
  print('Фраза 2 длиннее фразы 1')
else: print('Фразы равной длины')
```

**Задание 6 (необязательное)**

Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник). Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

* если пользователь выбрал круг, запрашиваем его радиус,
* если треугольник – длины трех его сторон;
* если прямоугольник – длины двух его сторон.

```python
import math
print('Введите тип фигуры:')
figure = str(input())
if figure == 'Круг':
  print('Введите радиус круга:')
  radius = int(input())
  print('Площадь круга: ', round(math.pi*radius**2, 2))
elif figure == 'Треугольник':
  print('Введите длину стороны A:')
  a = int(input())
  print('Введите длину стороны В:')
  b = int(input())
  print('Введите длину стороны С:')
  c = int(input())
  p = (a+b+c)/2 # Полупериметр треугольника
  print('Площадь треугольника: ', round(math.sqrt(p*(p-a)*(p-b)*(p-c)), 2))
  # https://2mb.ru/matematika/geometriya/ploshhad-treugolnika-po-trem-storonam/ Формула Герона
elif figure == 'Прямоугольник':
  print('Введите длину стороны A:')
  a = int(input())
  print('Введите длину стороны В:')
  b = int(input())
  print('Площадь прямоугольника: ', round(a*b, 2))
else: print('Что-то пошло не так!')
```
