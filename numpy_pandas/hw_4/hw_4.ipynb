**Задание 1**

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

- если источник traffic_source равен yandex или google, то в source_type ставится organic 
- для источников paid и email из России - ставим ad
- для источников paid и email не из России - ставим other
- все остальные варианты берем из traffic_source без изменений

```python
import pandas as pd
```

```python
log = pd.read_csv('visit_log.csv', sep=';')
log.head()
```

```python
def classification(df):
    """ Разделение traffic_source на типы (source_type) """
    if df['traffic_source'] in ['google', 'yandex']:
        return 'organic'
    elif (df['traffic_source'] in ['paid', 'email']) and (df['region']=='Russia'):
        return 'ad'
    elif (df['traffic_source'] in ['paid', 'email']) and (df['region']!='Russia'):
        return 'other'
    return df['traffic_source']
```

```python
# Способ №1
log['source_type'] = log.apply(classification, axis=1)
log.head(20)
```

```python
# Способ №2
log['source_type'] = log['traffic_source']
log.loc[log['traffic_source'].str.contains('google|yandex'), 'source_type'] = 'organic'
log.loc[(log['traffic_source'].str.contains('paid|email'))&(log['region'].str.contains('Russia')), 'source_type'] = 'ad'
log.loc[(log['traffic_source'].str.contains('paid|email'))&(~log['region'].str.contains('Russia')), 'source_type'] = 'other'
log.head(20)
```

**Задание 3**

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
max_time_range = ratings.groupby('userId').agg({'timestamp':['min', 'max'], 'userId':['count']})
max_time_range = max_time_range[max_time_range['userId']['count']>100]
max_time_range['diff_time'] = max_time_range['timestamp']['max']-max_time_range['timestamp']['min']
max_time_range['range_time_hours'] = pd.to_datetime(max_time_range['diff_time'], unit='s').dt.hour
max_time_range[['userId', 'range_time_hours']].head()
# Как можно переименовать колонку с двойным/двухэтажным названием?
# https://pandas.pydata.org/pandas-docs/stable/whatsnew/v0.20.0.html#deprecate-groupby-agg-with-a-dictionary-when-renaming
# https://github.com/pandas-dev/pandas/issues/32156
```
