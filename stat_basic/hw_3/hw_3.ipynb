Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш.

* 777: 200 руб.

* 999: 100 руб.

* 555: 50 руб.

* 333: 15 руб.

* 111: 10 руб.

* *77: 5 руб.

* **7: 3 руб.

* *00: 2 руб.

* **0: 1 руб.

Вопросы:

1. Выгодна ли игра игроку?
2. Сколько в среднем приобретает или теряет игрок за одну игру?

```python
def check_win(drop_number):
    if drop_number==777:
        return 200
    elif drop_number==999:
        return 100
    elif drop_number==555:
        return 50
    elif drop_number==333:
        return 15
    elif drop_number==111:
        return 10
    elif drop_number % 100 ==77:
        return 5
    elif drop_number % 10 ==7:
        return 3
    elif drop_number % 100 ==0:
        return 2
    elif drop_number % 10 ==0:
        return 1
    else:
        return 0
```

```python
import numpy as np
win_player = 0
number_games = 1000000
for i in range(number_games):
    drop_number = np.random.randint(0,1000)
    win_player += check_win(drop_number)
win_player/number_games
print(f'Средний выигрыш с игры: {round(win_player/number_games, 2)} руб.')
```

1. Нет игра не выгодна. 
1. При достаточно долгой игре убыток для игрока будет составлять порядка 20% от вложенных средств, т.е. игрок в среднем теряет 20 копеек с каждой игры.
