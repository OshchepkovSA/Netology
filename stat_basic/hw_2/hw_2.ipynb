```python
# Задача. Сделаем единый датафрейм из представленных данных, чтобы удобнее было анализировать
# 1. Добавим в каждый датафрейм столбец year
# 2. Определим одинаковые столбцы для всех таблиц и переименуем их, приведя заголовки к одному виду
# 3. Соеденим датафреймы по одинаковым столбцам, способом union (как в sql)
```

```python
h2015['year'] = 2015
h2016['year'] = 2016
h2017['year'] = 2017
h2018['year'] = 2018
h2019['year'] = 2019
```

```python
# rank 
h2015.rename(columns={h2015.columns[2]: "rank"}, inplace = True)
h2016.rename(columns={h2016.columns[2]: "rank"}, inplace = True)
h2017.rename(columns={h2017.columns[1]: "rank"}, inplace = True)
h2018.rename(columns={h2018.columns[0]: "rank"}, inplace = True)
h2019.rename(columns={h2019.columns[0]: "rank"}, inplace = True)
# country
h2015.rename(columns={h2015.columns[0]: "country"}, inplace = True)
h2016.rename(columns={h2016.columns[0]: "country"}, inplace = True)
h2017.rename(columns={h2017.columns[0]: "country"}, inplace = True)
h2018.rename(columns={h2018.columns[1]: "country"}, inplace = True)
h2019.rename(columns={h2019.columns[1]: "country"}, inplace = True)
# score
h2015.rename(columns={h2015.columns[3]: "score"}, inplace = True)
h2016.rename(columns={h2016.columns[3]: "score"}, inplace = True)
h2017.rename(columns={h2017.columns[2]: "score"}, inplace = True)
h2018.rename(columns={h2018.columns[2]: "score"}, inplace = True)
h2019.rename(columns={h2019.columns[2]: "score"}, inplace = True)
# gdp_per_capita
h2015.rename(columns={h2015.columns[5]: "gdp_per_capita"}, inplace = True)
h2016.rename(columns={h2016.columns[6]: "gdp_per_capita"}, inplace = True)
h2017.rename(columns={h2017.columns[5]: "gdp_per_capita"}, inplace = True)
h2018.rename(columns={h2018.columns[3]: "gdp_per_capita"}, inplace = True)
h2019.rename(columns={h2019.columns[3]: "gdp_per_capita"}, inplace = True)
# social_support	
h2015.rename(columns={h2015.columns[6]: "social_support"}, inplace = True)
h2016.rename(columns={h2016.columns[7]: "social_support"}, inplace = True)
h2017.rename(columns={h2017.columns[6]: "social_support"}, inplace = True)
h2018.rename(columns={h2018.columns[4]: "social_support"}, inplace = True)
h2019.rename(columns={h2019.columns[4]: "social_support"}, inplace = True)
#  lifespan
h2015.rename(columns={h2015.columns[7]: "lifespan"}, inplace = True)
h2016.rename(columns={h2016.columns[8]: "lifespan"}, inplace = True)
h2017.rename(columns={h2017.columns[7]: "lifespan"}, inplace = True)
h2018.rename(columns={h2018.columns[5]: "lifespan"}, inplace = True)
h2019.rename(columns={h2019.columns[5]: "lifespan"}, inplace = True)
# freedom
h2015.rename(columns={h2015.columns[8]: "freedom"}, inplace = True)
h2016.rename(columns={h2016.columns[9]: "freedom"}, inplace = True)
h2017.rename(columns={h2017.columns[8]: "freedom"}, inplace = True)
h2018.rename(columns={h2018.columns[6]: "freedom"}, inplace = True)
h2019.rename(columns={h2019.columns[6]: "freedom"}, inplace = True)
# generosity
h2015.rename(columns={h2015.columns[10]: "generosity"}, inplace = True)
h2016.rename(columns={h2016.columns[11]: "generosity"}, inplace = True)
h2017.rename(columns={h2017.columns[9]: "generosity"}, inplace = True)
h2018.rename(columns={h2018.columns[7]: "generosity"}, inplace = True)
h2019.rename(columns={h2019.columns[7]: "generosity"}, inplace = True)
# corruption
h2015.rename(columns={h2015.columns[9]: "corruption"}, inplace = True)
h2016.rename(columns={h2016.columns[10]: "corruption"}, inplace = True)
h2017.rename(columns={h2017.columns[10]: "corruption"}, inplace = True)
h2018.rename(columns={h2018.columns[8]: "corruption"}, inplace = True)
h2019.rename(columns={h2019.columns[8]: "corruption"}, inplace = True)
h2019.columns
```

```python
happy = pd.concat([h2015,h2016,h2017,h2018,h2019])
```

```python
happy = happy[['rank', 'country', 'score', 'gdp_per_capita', 'social_support', 'lifespan', 'freedom', 'generosity', 'corruption', 'year']]
happy.head()
```

```python
# Проверяем тип данных столбцов
happy.info()
```

```python
# Переименуем страны
happy['country'] = happy['country'].str.capitalize()
happy['country'].value_counts()
```

```python
# Для простоты уберем страны, которые встречаются меньше 5 раз/лет
df = happy.groupby('country').count()
df = df[df['rank']==5]
country_5 = list(df.index.values)
happy = happy[happy['country'].isin(country_5)]
happy['country'].value_counts()
```

```python
happy.head()
```

```python
# Ярко выраженую положительную корреляцию со score иемеют оценочные показателя gdp_per_capita, social_support, lifespan
# Разумно предположить, что ощущение высокой удовлетворенности человек испытывает:
# при высоком уровне благосостояния gdp_per_capita,
# при доступности и понятности мер помощи и поддержки в беде или сложной жизненной ситуацию social_support

sns.pairplot(happy[['rank',	'score', 'gdp_per_capita', 'social_support', 'lifespan', 'freedom', 'generosity', 'corruption']])
```

```python
# При существенно высоком ВВП происходит резкий рост размаха corruption в странах
# При высоких свободах наблюдается высокий разброс corruption, при низких свободах разброс corruption незначителен, и сама corruption низкая

# Можно предположить, что ощущение свободы достигается уменьшением контроля со стороны государства за сферами жизнедеятельности человека
# и при значительном ослаблении контроля возможен скачек corruption совместно с ощущением свободы. Что может натолкнуть на мысль поиска
# "золотой середины" уровня контроля в государстве
```

```python
# Выделим из подготовленного сводного датафрейма таблица по годам
happy_15 = happy[happy['year']==2015]
happy_16 = happy[happy['year']==2016]
happy_17 = happy[happy['year']==2017]
happy_18 = happy[happy['year']==2018]
happy_19 = happy[happy['year']==2019]
```

```python
# Сделаем датафрейм по странам со средними значениями показателей
#  ценность представляют - 'score', 'gdp_per_capita', 'social_support', 'lifespan', 'freedom', 'generosity', 'corruption'
df = happy.groupby('country').mean().reset_index()

df.head()
```

```python
# Визуализируем разброс средних оценочных показателей по странам
df[['gdp_per_capita',	'social_support',	'lifespan',	'freedom',	'generosity',	'corruption']].plot(kind='box',
                    title='Разброс оценочных показатели')
```

```python
country_year_gdp = happy[happy['country'].isin(happy_15_.index.values)].pivot_table(
                        index='country', 
                        columns='year', 
                        values='gdp_per_capita', 
                        aggfunc=sum).fillna(0)

country_year_gdp
```

```python
# Можно заметить равномерный рост ВВП на душу населения в первой 10 стран по gdp_per_capita от 2015 года
# начиная с 2015 по 2017 год включительно и сушественный спад показателя gdp_per_capita в 2018 год, 
# исключение ОАЭ у которой, наоборот произошел резкий скачек gdp_per_capita
# Могу предположить, что в 2018 году произошел некоторый кризис и резкий спад в показателях мировой экономики.
# которые и привели к падению показателя gdp_per_capita
sns.heatmap(country_year_gdp,1)
```

```python
happy_15_ = happy_15.set_index('country')
happy_15_ = happy_15_[['gdp_per_capita']].sort_values(by='gdp_per_capita',ascending=False).head(10)
country_list = list(happy_15_.index.values)
happy_19_ = happy_19[happy_19['country'].isin(country_list)]
happy_19_ = happy_19_[['country','gdp_per_capita']].set_index('country')
happy_19_
```

```python

# libraries
import numpy as np
import matplotlib.pyplot as plt
 
# width of the bars
barWidth = 0.3
 
# Choose the height of the blue bars
bars1 = happy_15_['gdp_per_capita']
 
# Choose the height of the cyan bars
bars2 = happy_19_['gdp_per_capita']
 
# The x position of bars
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
 
# Create blue bars
plt.bar(r1, bars1, width = barWidth, color = 'blue', edgecolor = 'black', capsize=7, label='2015')
 
# Create cyan bars
plt.bar(r2, bars2, width = barWidth, color = 'cyan', edgecolor = 'black', capsize=7, label='2019')
 
# general layout
plt.xticks([r + barWidth for r in range(len(bars1))], country_list)
plt.xticks(rotation=90)
plt.ylabel('ВВП')
plt.legend()
 
# Show graphic
plt.show()
```
